```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./inputs/Environmental_Data_Deep_Moor_2015.csv')
```

```python
def hours_dict(date):
    day = df[df['date']==date]
    hours = [time.split(':')[0] for time in day['time']]
    hours_dict = {i:hours.count(str(i)) for i in np.arange(24)}
    return hours_dict
```

```python
feb_15 = hours_dict('2015_02_15')
plt.plot(feb_15.keys(), feb_15.values())
plt.show()
```

```python
feb_15 = hours_dict('2015_02_15')
feb_16 = hours_dict('2015_02_16')
plt.plot(feb_15.keys(), feb_15.values())
plt.plot(feb_16.keys(), feb_16.values())
plt.show()
```

```python
feb_15 = hours_dict('2015_02_15')
feb_16 = hours_dict('2015_02_16')
plt.plot(feb_15.keys(), feb_15.values())
plt.plot(feb_16.keys(), feb_16.values())
plt.xticks((np.arange(0,24,6)))

plt.xlim(-3,26)
plt.xlabel("Hours of the day")

plt.ylim(-3,28)
plt.ylabel("Measurements taken")
plt.title("Measurements by day")
plt.show()
```
